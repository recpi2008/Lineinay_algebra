1. Решить систему уравнений методом Гаусса:

$$\begin{cases}
x_{1}+x_{2}-x_{3}-2x_{4}=0, \\
2x_{1}+x_{2}-x_{3}+x_{4}=-2, \\
x_{1}+x_{2}-3x_{3}+x_{4}=4.
\end{cases}$$
Преобразуем расширенную матрицу системы путем элементарных преобразований:

$$\begin{pmatrix}
\left.\begin{matrix}
1 &amp; 1 &amp; -1 &amp; -2 \\
2 &amp; 1 &amp; -1 &amp; 1 \\
1 &amp; 1 &amp; -3 &amp; 1
\end{matrix}\right|
\begin{matrix}
0\\
-2\\
4
\end{matrix}
\end{pmatrix}$$
Отнимем из второй строки первую, умноженную на 2:

$$\begin{pmatrix}
\left.\begin{matrix}
1 &amp; 1 &amp; -1 &amp; -2\\
0 &amp; -1 &amp; 1 &amp;5\\
1 &amp; 1 &amp; -3 &amp; 1
\end{matrix}\right|
\begin{matrix}
0\\
-2\\
4
\end{matrix}
\end{pmatrix}$$
Отнимем из третьей строки первую:

$$\begin{pmatrix}
\left.\begin{matrix}
1 &amp; 1 &amp; -1 &amp; -2\\
0 &amp; -1 &amp; 1 &amp; 5\\
0 &amp; 0 &amp; -2 &amp; 3
\end{matrix}\right|
\begin{matrix}
0\\
-2\\
4
\end{matrix}
\end{pmatrix}$$
Система имеет бесконечное кол-во решений.

Поделим третью строку на 3:

$$\begin{pmatrix}
\left.\begin{matrix}
1 &amp; 1 &amp; -1 &amp; -2\\
0 &amp; -1 &amp; 1 &amp; 5\\
0 &amp; 0 &amp; -\frac{2}{3} &amp; 1
\end{matrix}\right|
\begin{matrix}
0\\
-2\\
\frac{4}{3}
\end{matrix}
\end{pmatrix}$$
Оставшаяся матрица соответствует системе:

$$\begin{cases}
x_{1} + x_{2} - x_{3} -2 \cdot x_{4} = 0, \\
-x_{2} + x_{3} + 5 \cdot x_{4} = -2, \\
-\frac{2}{3} \cdot x_{3} + x_{4} = \frac{4}{3}
\end{cases}$$
Бесконечное множество решений принято записывать в виде общего решения.

Последнее уравнение содержит две неизвестных. В данном случае можно рассматривать в качестве свободного параметра $x_{4}=c$. Тогда, выражая остальные переменные через $c$, получим:

$x_{4}=c,$

$x_{3}=\frac{4-3c}{-2} \;$

$x_{2}=2 + 5 \cdot c + \frac{4-3c}{-2},$

$x_{1}=-2 -3 \cdot c.$

Таковым будет общее решение системы. Подставляя произвольные числа вместо $c$, мы получим частное решение. Например, при $c=0$:

$x_{1}=-2,$ $x_{2}=0,$ $x_{3}=-2,$ $x_{4}=0.$

Параметр $c$ может принимать бесконечное количество значений, при которых уравнения в системе будут обращаться в тождества.

5. Решить систему уравнений методом Крамера: а) $\begin{cases}
x_{1}-2x_{2}=1 \\
3x_{1}-4x_{2}=7
\end{cases}$

Решение 5а):

Найдем определитель матрицы:

$detA=\begin{vmatrix}
1 &amp; -2\\
3 &amp; -4\\
\end{vmatrix}=(1 \cdot -4) - (-2 \cdot 3) = 2$

Найдем определитель $detA_{1}$:

$detA_{1}=\begin{vmatrix}
1 &amp; -2\\
7 &amp; -4\\
\end{vmatrix}=(1 \cdot -4) - (-2 \cdot 7) = 10$

Найдем определитель $detA_{2}$:

$detA_{2}=\begin{vmatrix}
1 &amp; 1\\
3 &amp; 1\\
\end{vmatrix}=(1 \cdot 7) - (1 \cdot 3) = 4$

Найдем значения $x_{1}, x_{2}$:

$x_{1}=10/2=5$

$x_{2}=4/2=2$

Проверим полученное решение:

$\begin{cases}
5-4=1 \\
15-8=7
\end{cases}$
